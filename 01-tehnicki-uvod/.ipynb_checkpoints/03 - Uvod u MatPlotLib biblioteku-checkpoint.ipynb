**MatPlotLib** je Python biblioteka koja se koristi za vizelizaciju podataka. Na [ovom](https://matplotlib.org/) linku se nalazi zvanicna stranica, a [ovde](https://matplotlib.org/users/tutorials.html) se moze pronaci koristan tutorijal.

**PyPlot** je modul MatPlotLib biblioteke koji pruza podrsku za iscrtavanje grafika poput MATLAB-a. Svaki poziv funkcije vrsi odredjenu promenu na figuri za plot-ovanje: npr. kreira figuru i povrsinu za iscrtavanje grafika, iscrtava neke linije u oblasti za crtanje, dodaje figuri neke oznake, itd.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#da bi se matplotlib grafovi prikazali u jupyter notebook-u odmah nakon naredbe 
#(kod novijih verzija Jupyter Notebook-a ne mora, vec je u konfiguracijama)        
%matplotlib inline
```

# Grafik funkcije

**Grafik funkcije** se koristi za vizuelni prikaz ponasanja funkcije na nekom intervalu.

```python
x = np.linspace(-np.pi, np.pi, 100)
sin_x = np.sin(x)
cos_x = np.cos(x)

plt.plot(x, sin_x, label='sin', color='blue')
plt.plot(x, cos_x, label='cos', color='green')
plt.legend(loc='upper left')
plt.show()
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
sin_x = np.sin(x)

plt.plot(x, sin_x, linestyle='--', linewidth=2)
plt.title('Grafik sinusne funkcije')
plt.xlabel('Domen funkcije')
plt.ylabel('Vrednost funkcije')
plt.xticks(np.linspace(-7, 7, 15))
plt.yticks(np.linspace(-1, 1, 9))
plt.show()
```

# Dijagram rasipanja (eng. scatter plot)

**Dijagram rasipanja** se koristi za vizuelizaciju zavisnosti dve (slucajne) promeljive. U kontekstu masinskog ucenja koristi se za uvid u zavisnosti medju podacima (atributima).

```python
x = np.linspace(0, 10, 20)
y = np.linspace(0, 5, 20)

plt.plot(x, y, 'o')
plt.title('Linearna zavisnost medju podacima')
plt.show()
```

```python
x = np.random.randint(0, 10, 10)
y = np.random.randint(0, 20, 10)

plt.scatter(x, y, marker='v', color='green')
plt.title('Nema zavisnosti medju podacima')
plt.show()
```

# Kruzni grafikon (eng. pie chart)

**Kruzni grafikon** se koristi za vizuelni prikaz kako je neki ukupan iznos podeljen izmedju razlicitih kategorija. U kontekstu masinskog ucenja koristi se za vizuelizaciju udela pojedinacnih kategorija (vrednosti kategoricke slucajne promenljive) u podacima.

```python
kategorije = ['ZA', 'PROTIV', 'UZDRZANO']
glasovi = [450, 209, 331]

plt.pie(glasovi, labels=kategorije, autopct='%1.1f%%')
plt.title('Rezultati ankete')
plt.show()
```

# Stubicasti grafikon (eng. bar chart)

**Stubicasti grafikon** se koristi za vizuelni prikaz numerickih vrednosti (najcesce frekvencija) pridruzenih razlicitim kategorijama. U kontekstu masinskog ucenja koristi se za vizuelizaciju raspodele kategorickih atributa (kategorickih slucajnih promenljiva) u podacima.

```python
#liste kategorickih vrednosti
boje = ['Crvena', 'Plava', 'Zelena', 'Zuta', 'Roze']
frekvencija = [8, 7, 5, 9, 6]

plt.bar(boje, frekvencija)
plt.xlabel('Boja')
plt.ylabel('Frekvencija')
plt.title('Omiljena boja na uzorku od 50 djaka')
plt.show()
```

# Histogram

**Histogram** se koristi za vizuelni prikaz numerickog skupa vrednosti grupisanjem bliskih vrednosti u zajednicke korpice (eng. bins). U kontekstu masinskog ucenja koristi se za vizuelizaciju (aproksimacije) raspodele numerickih atributa (kontinualnih slucajnih promenljivih) u podacima.

```python
x = np.random.randn(100)

plt.hist(x, bins=8)
plt.xlabel('Vrednosti')
plt.ylabel('Frekvencije')
plt.title('Random vrednosti iz normalne raspodele N(0, 1)') 
plt.show()
```

**NAPOMENA**: Primetimo jos da funkcije <code>bar()</code> i <code>hist()</code> ocekuju podatke u razlicitom obliku - funkciji <code>bar()</code> se prosledjuju unapred izracunate frekvencije, dok se funkciji <code>hist()</code> prosledjuju 'sirovi' podaci i ona sama izracunava frekvencije!

Sledeci primeri pokazuju da se histogram moze koristiti gde i stubicasti grafikon (kategorijama ce biti dodeljene redom celobrojne vrednosti), i da se stubicasti grafikon moze koristiti gde i histogram uz diskretizaciju podataka:

```python
x = ['plava', 'zuta', 'plava', 'plava', 'zelena', 'plava', 'roza', 'zelena', 'zelena', 'zuta']

plt.hist(x)
plt.show()
```

```python
plt.figure(figsize=(8, 5))

#liste kategorickih vrednosti
osobe = ['150-','[150, 160]', '[160, 170]', '[170, 180]', '[180, 190]', '[190, 200]', '200+']
visina = [3, 15, 25, 33, 14, 8, 1]

plt.bar(osobe, visina)
plt.xlabel('CM')
plt.ylabel('Frekvencija')
plt.title('Visina na uzorku od 100 studenata')
plt.show()
```

# Toplotna mapa (eng. heat map)

Prikaz podataka u formi **toplotne mape** se koristi za vizuelizaciju trojki podataka (npr. temperatura vazduha u odnosu na jacinu vetra i oblacnost). **Toplotna mapa** prikazuje vrednost za promenljivu od interesa preko dve osne promenljive kao mrezu obojenih kvadrata, pri cemu intenzitet boje celije oslikava vrednost promenljive od interesa (obicno tamnije nijanse predstavljaju vece vrednosti od svetlije). U kontekstu masinskog ucenja koristi se za vizuelizaciju jacina korelacija izmedju atributa (slucajnih promenljivih) u podacima.

```python
A = np.random.randint(0, 100, size=(2, 2))

plt.imshow(A)              #II nacin: plt.matshow()  
plt.colorbar()
plt.xticks([0, 1])
plt.yticks([0, 1])

for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        plt.text(i, j, A[i, j])

plt.show()
```

# Grafik kontura (eng. contour plot)

**Grafik kontura** je dvodimenzionalni grafik funkcije od dve promenljive koji prikazuje konturne linije (izolinije) funkcije - linije koje spajaju tacke u kojima funkcija uzima istu vrednost. U kontekstu masinskog ucenja koristi se za vizuelizaciju kontura funkcije greske ili kod problema klasifikacije za vizuelizaciju granice izmedju klasa.

```python
N = 30
x = np.linspace(-6.0, 6.0, N)
y = np.linspace(-6.0, 6.0, N)

X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)       

plt.figure(figsize=(5, 4))
plt.contourf(X, Y, Z)
plt.colorbar()
plt.show()
```

Konture se mogu videti, ali ne toliko jasno, i pomocu toplotne mape:

```python
plt.imshow(Z)
plt.colorbar()
plt.xticks(np.arange(0, 30, 5), np.arange(-6, 6, 2))
plt.yticks(np.arange(0, 30, 5), np.arange(-6, 6, 2))
plt.show()
```

# 3D dijagram rasipanja (eng. 3D scatter plot)

**3D dijagram rasipanja** je trodimenzionalni grafik funkcije od dve promenljive koji prikazuje vrednosti funkcije u pojedinim tackama. U kontekstu masinskog ucenja se koristi, kao i grafik kontura, kod problema klasifikacije za vizuelizaciju granice izmedju klasa ili kod ucenja reprezentacije podataka da bi se prikazalo kako instance ispunjavaju prostor atributa. 

Matplotlib was initially designed with only two-dimensional plotting in mind. Around the time of the 1.0 release, some three-dimensional plotting utilities were built on top of Matplotlib's two-dimensional display, and the result is a convenient (if somewhat limited) set of tools for three-dimensional data visualization. 3D plots are enabled by importing the **mplot3d** toolkit. Once this submodule is imported, a three-dimensional axes can be created by passing the keyword **projection='3d'** to any of the normal axes creation routines.

```python
from mpl_toolkits.mplot3d import Axes3D      
```

```python
N = 30
x = np.linspace(-6.0, 6.0, N)
y = np.linspace(-6.0, 6.0, N)

X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)  

x_flatten = X.flatten()
y_flatten = Y.flatten()
z_flatten = Z.flatten()

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")                  #kreiranje objeka klase Axes3D  
ax.scatter3D(x_flatten, y_flatten, z_flatten)    #object-oriented interface gives more control for customizing plot
ax.set_xlabel('X', color='red')                  #in comparison to functional interface 
ax.set_ylabel('Y', color='red')
ax.set_zlabel('Z', color='red')

plt.show()
```

# Iscrtavanje vise grafika na jednoj figuri

Funkcija <code>subplot()</code> dodaje podgraf na odgovarajucoj poziciji u okviru tekuce figure za plot-ovanje. Kao argument se zadaje uredjena trojka <code>(nrovs, ncols, index)</code> ili trocifreni broj cije cifre stotina, desetica i jedinica redom odgovaraju <code>nrovs</code>, <code>ncols</code> i <code>index</code>. Podgraf koji se dodaje ce zauzeti polozaj na indeksu <code>index</code> (enumeracija krece od 1) u okviru mreze (eng. grid) dimenzija <code>nrows</code> $\times$ <code>ncols</code>.

```python
plt.figure(figsize=(10, 5))

#-----------------------------------------------------
plt.subplot(121)

x = np.random.randint(0, 100, size=(100, 2))
y = np.random.randint(0, 2, size=100)
class_color = ['blue', 'red']
colors = [class_color[y_i] for y_i in y]

plt.scatter(x[:, 0], x[:, 1], color=colors)

#-----------------------------------------------------
plt.subplot(122)

klase = ['klasa 0', 'klasa 1']
frekvencije = [np.sum(x==0), np.sum(x==1)]

plt.bar(klase, frekvencije)
```
